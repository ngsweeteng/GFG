```python
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
# Read dataset

rows_to_keep = range(2,21)
df = pd.read_csv('GFG Data Analyst test.csv', skiprows = lambda x: x not in rows_to_keep, usecols = range(0,7))
```

```python
# Clean the dataset

# 1) Remove commas in the numbers
df = df.replace(',','', regex=True)

# 2) Change the data types of Visits, Orders and Revenue in $ to int64
df['Visits'] = pd.to_numeric(df['Visits'])
df['Orders'] = pd.to_numeric(df['Orders'])
df['Revenue in $'] = pd.to_numeric(df['Revenue in $'])
```

```python
# Double check on the data types

df.info()
```

```python
# Add in the different metrics

df['CR'] = df['Orders'] / df['Visits'] * 100
df['ABS'] = df['Revenue in $'] / df['Orders']
df['CIR'] = df['Spend in $'] / df['Revenue in $'] * 100
```
